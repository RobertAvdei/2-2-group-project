```python
import os
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split

# -------- CONFIG --------
SOURCE_DIR = "all_data/training_data"     # folder with all images
OUTPUT_DIR = "all_data"    # output folder
TRAIN_RATIO = 0.8
RANDOM_STATE = 42
IMG_EXTENSIONS = (".jpg", ".jpeg", ".png")
# ------------------------

data = []

# Collect filenames and labels
for filename in os.listdir(SOURCE_DIR):
    if filename.lower().endswith(IMG_EXTENSIONS):
        label = filename.rsplit("_", 1)[0]
        data.append({"filename": filename, "label": label})

df = pd.DataFrame(data)

display(df)
```

```python
# Stratified split
train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["label"],
    random_state=RANDOM_STATE
)

def move_files(df_split, split_name):
    for _, row in df_split.iterrows():
        src_path = os.path.join(SOURCE_DIR, row["filename"])
        dst_dir = os.path.join(OUTPUT_DIR, split_name, row["label"])
        dst_path = os.path.join(dst_dir, row["filename"])

        os.makedirs(dst_dir, exist_ok=True)
        shutil.move(src_path, dst_path)

# Move files
move_files(train_df, "train")
move_files(val_df, "val")

print("âœ… Dataset split complete!")
print(f"Training images: {len(train_df)}")
print(f"Validation images: {len(val_df)}")
```
